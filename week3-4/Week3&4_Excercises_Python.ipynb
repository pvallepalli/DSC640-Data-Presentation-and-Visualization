## Python Excercises

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('world-population.xlsm')
```

### Python - Line Chart

```python
plt.plot(df['Year'], df['Population'])
plt.title('Python - Line chart of World population ')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

### Python - Step Chart

```python
plt.step(df['Year'], df['Population'])
plt.title('Python Step chart of World population ')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```
