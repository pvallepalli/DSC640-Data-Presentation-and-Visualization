## R Excercises

```r
# Import required packages
library('magrittr')
library("ggplot2")
library("dplyr")
library("xlsx")
```

```r
file = paste(getwd(), '/world-population.xlsm', sep = '')
df = xlsx::read.xlsx(file, sheetIndex=1, stringsAsFactors=FALSE)
```

### R - Line Chart

```r
ggplot2::ggplot(data=df, ggplot2::aes(x=Year, y=Population)) + ggplot2::geom_line(linetype='solid' ,color='blue', size=1.0) + ggplot2::ggtitle(label='Line chart in R for Population growth')
```

### R - Step Chart

```r
ggplot2::ggplot(data=df, ggplot2::aes(x=Year, y=Population)) + ggplot2::geom_step(linetype='solid', color='blue', size=1.0) + ggplot2::ggtitle(label='Step chart in R for Population growth')
```
