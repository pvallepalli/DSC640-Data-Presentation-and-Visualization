# Python Excercises

```python
import pandas as pd
```

```python
df1 = pd.read_excel('obama-approval-ratings.xls')
```

## Bar Chart - Python

```python
import matplotlib.colors
 
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["green","red","grey"])
 
# Horizontal Bar Plot
df1.plot(x='Issue',
        kind='bar',
        stacked=False,
        width=0.75,
        title='Bar chart for approval rating using Python', figsize=(12,8), colormap=cmap)
```

## Stacked Bar Chart - Python

```python
import matplotlib.colors
 
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["green","red","grey"])
 
# Stacked Bar chart
df1.plot(x='Issue',
        kind='bar',
        stacked=True,
        width=0.75,
        title='Stacked Bar chart for approval rating using Python', figsize=(12,8), colormap=cmap)
```

## Pie Chart - Python

```python
plt = df1.plot.pie(y='Approve', labels=df1['Issue'], title="Pie chart of approval by Issue using python", startangle=90, figsize=(14,10), autopct='%1.1f%%', legend=False)
```

## Donut chart - Python

```python
import matplotlib.pyplot as plt

plt.pie(df1.Disapprove, labels=df1.Issue, autopct = '%1.1f%%', pctdistance = 0.75)
circle = plt.Circle((0,0), 0.40, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(circle)
fig.set_size_inches(12, 8)

plt.title("Donut chart for disapproval percentage by Issue")
# Show compact plot
plt.tight_layout()
plt.show()
```
