# R Plots

```r
# Import required packages
library('magrittr')
library("ggplot2")
library("dplyr")
```

```r
file = paste(getwd(), '/expenditures.txt', sep = '')
df1 = read.table(file, header = TRUE, sep = '\t', dec = '.', fill = TRUE)
```

```r
df = read.csv("unemployement-rate-1948-2010.csv", sep=',', stringsAsFactors = FALSE) %>% 
    dplyr::mutate(Value = as.numeric(Value)) %>% 
    as.data.frame()
```

```r
print(head(df))
```

## R: Tree Map

```r
treemap::treemap(df1,index = c('category'),
                 vSize = 'expenditure',
                 title = 'Expenditure by Category', 
                 palette = 'RdYlGn')
```

## R: Area Plot

```r
avg_unemployement = df %>%
    dplyr::group_by(Year) %>%
    dplyr::summarize('Average Value' = mean(Value))

head(avg_unemployement)
```

```r
ggplot2::ggplot(avg_unemployement, ggplot2::aes(x=Year , y=`Average Value`)) +
ggplot2::geom_area( fill='blue', alpha=.6) +
ggplot2::geom_line() +
ggplot2::ggtitle('Average Value by Year')
```

## R: Stacked Area Plot 

```r
ggplot2::ggplot(df1, ggplot2::aes(x=year, y=expenditure, fill=category)) + ggplot2::geom_area() + ggplot2::ggtitle('Expenditure of each category by Year')
```
